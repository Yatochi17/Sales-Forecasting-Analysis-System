# 2. Forecast Specific Product Demands (LTSM)

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates # Import the mdates module
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import Dense, LSTM
from sklearn.model_selection import train_test_split
```

```python
# Load the dataset
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Data Analytics & Machine Learning/Assignment/ds_merged(2.0).csv')
```

```python
# Handling missing values by filling with "null" for numerical and "none" for categorical columns
for column in df.columns:
    if df[column].dtype == np.number:
        df[column].fillna("Null", inplace=True)
    else:
        df[column].fillna("None", inplace=True)
```

```python
# Preprocess the data
df['Created Time'] = pd.to_datetime(df['Created Time'])
df.set_index('Created Time', inplace=True)

# Resample the data by day and sum the quantities
df_resampled = df['Quantity'].resample('D').sum().fillna(0)

# Plot the demand to visualize the data
plt.figure(figsize=(12, 6))
plt.plot(df_resampled, label='Demand')
plt.title('Product Demand Over Time')
plt.xlabel('Date')
plt.ylabel('Demand')
plt.legend()
plt.show()

# Scale the data
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(df_resampled.values.reshape(-1, 1))

# Prepare the data for LSTM
def create_dataset(data, time_step=1):
    X, Y = [], []
    for i in range(len(data) - time_step - 1):
        a = data[i:(i + time_step), 0]
        X.append(a)
        Y.append(data[i + time_step, 0])
    return np.array(X), np.array(Y)

time_step = 10
X, Y = create_dataset(scaled_data, time_step)

# Reshape the input to be [samples, time steps, features]
X = X.reshape(X.shape[0], X.shape[1], 1)

# Build and train the LSTM model
model = Sequential()
model.add(LSTM(50, return_sequences=True, input_shape=(time_step, 1)))
model.add(LSTM(50, return_sequences=False))
model.add(Dense(25))
model.add(Dense(1))

model.compile(optimizer='adam', loss='mean_squared_error')
model.fit(X, Y, batch_size=1, epochs=1)

# Split the data into training and testing sets
train_size = int(len(scaled_data) * 0.8)
test_size = len(scaled_data) - train_size
train_data, test_data = scaled_data[0:train_size, :], scaled_data[train_size:len(scaled_data), :]

X_train, Y_train = create_dataset(train_data, time_step)
X_test, Y_test = create_dataset(test_data, time_step)

X_train = X_train.reshape(X_train.shape[0], X_train.shape[1], 1)
X_test = X_test.reshape(X_test.shape[0], X_test.shape[1], 1)

# Make predictions
train_predict = model.predict(X_train)
test_predict = model.predict(X_test)

# Invert predictions
train_predict = scaler.inverse_transform(train_predict)
test_predict = scaler.inverse_transform(test_predict)
Y_train = scaler.inverse_transform([Y_train])
Y_test = scaler.inverse_transform([Y_test])

# Plot the results
train_predict_plot = np.empty_like(scaled_data)
train_predict_plot[:, :] = np.nan
train_predict_plot[time_step:len(train_predict) + time_step, :] = train_predict

test_predict_plot = np.empty_like(scaled_data)
test_predict_plot[:, :] = np.nan
test_predict_plot[len(train_predict) + (time_step * 2) + 1:len(scaled_data) - 1, :] = test_predict

plt.figure(figsize=(12, 6))
plt.plot(scaler.inverse_transform(scaled_data), label='True Demand')
plt.plot(train_predict_plot, label='Train Predict')
plt.plot(test_predict_plot, label='Test Predict')
plt.title('Product Demand Prediction')
plt.xlabel('Date')
plt.ylabel('Demand')
plt.legend()
plt.show()
```

```python
# Load the dataset
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Data Analytics & Machine Learning/Assignment/ds_merged(2.0).csv')
```

```python
# Handling missing values by filling with "null" for numerical and "none" for categorical columns
for column in df.columns:
    if df[column].dtype == np.number:
        df[column].fillna("Null", inplace=True)
    else:
        df[column].fillna("None", inplace=True)
```

```python
# Group by product and sum the quantities to find the top 10 products
top_products = df.groupby('Product Name')['Quantity'].sum().nlargest(10).index

# Filter the dataset to include only the top 10 products
top_products_df = df[df['Product Name'].isin(top_products)]

# Display the top products
print(top_products)
```

```python
# Group by product name and sum the quantities
product_quantities = df.groupby('Product Name')['Quantity'].sum()

# Sort the products by quantity in descending order and select the top 10
top_10_products = product_quantities.sort_values(ascending=False).head(10)

# Display the top 10 products
print(top_10_products)
```

```python
# Convert 'Created Time' to datetime type
df['Created Time'] = pd.to_datetime(df['Created Time'])
```

```python
# Group by product name and sum the quantities
top_10_products = df.groupby('Product Name')['Quantity'].sum().sort_values(ascending=False).head(10).index
top_products_df = df[df['Product Name'].isin(top_10_products)]

# Resample data by day and sum quantities for each product using 'Created Time'
product_demand = {}
for product in top_10_products:
    product_data = top_products_df[top_products_df['Product Name'] == product]
    product_data_resampled = product_data.resample('D', on='Created Time')['Quantity'].sum().fillna(0)
    product_demand[product] = product_data_resampled

# Prepare data for LSTM
def create_dataset(data, time_step=1):
    X, Y = [], []
    for i in range(len(data) - time_step - 1):
        a = data[i:(i + time_step), 0]
        X.append(a)
        Y.append(data[i + time_step, 0])
    return np.array(X), np.array(Y)

def prepare_lstm_data(data, time_step=10):
    scaler = MinMaxScaler(feature_range=(0, 1))
    scaled_data = scaler.fit_transform(data.values.reshape(-1, 1))
    X, Y = create_dataset(scaled_data, time_step)
    X = X.reshape(X.shape[0], X.shape[1], 1)
    return X, Y, scaler

time_step = 10
product_lstm_data = {}
for product, data in product_demand.items():
    X, Y, scaler = prepare_lstm_data(data, time_step)
    product_lstm_data[product] = (X, Y, scaler)

# Build LSTM model function
def build_lstm_model(input_shape):
    model = Sequential()
    model.add(LSTM(50, return_sequences=True, input_shape=input_shape))
    model.add(LSTM(50, return_sequences=False))
    model.add(Dense(25))
    model.add(Dense(1))
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model

# Train and predict for each product
product_predictions = {}

for product, (X, Y, scaler) in product_lstm_data.items():
    model = build_lstm_model((time_step, 1))
    model.fit(X, Y, batch_size=1, epochs=1)

    # Split data into training and testing sets
    train_size = int(len(X) * 0.8)
    X_train, X_test = X[:train_size], X[train_size:]
    Y_train, Y_test = Y[:train_size], Y[train_size:]

    # Make predictions
    train_predict = model.predict(X_train)
    test_predict = model.predict(X_test)

    # Invert predictions
    train_predict = scaler.inverse_transform(train_predict)
    test_predict = scaler.inverse_transform(test_predict)
    Y_train = scaler.inverse_transform([Y_train])
    Y_test = scaler.inverse_transform([Y_test])

    product_predictions[product] = (train_predict, test_predict, Y_train, Y_test)
```

```python
for product, data in product_demand.items():
    train_predict, test_predict, Y_train, Y_test = product_predictions[product]

    train_predict_plot = np.empty_like(data.values, dtype=float)
    train_predict_plot[:] = np.nan
    train_predict_plot[time_step:len(train_predict) + time_step] = train_predict.flatten()

    test_predict_plot = np.empty_like(data.values, dtype=float)
    test_predict_plot[:] = np.nan
    test_predict_plot[len(data) - len(test_predict):] = test_predict.flatten()

    plt.figure(figsize=(12, 6))
    plt.plot(data.index, data.values, label='True Demand')
    plt.plot(data.index, train_predict_plot, label='Train Predict')
    plt.plot(data.index, test_predict_plot, label='Test Predict')
    plt.title(f'Product Demand Prediction for {product}')
    plt.xlabel('Date')
    plt.ylabel('Demand')
    plt.legend()

    # Format the x-axis to show dates properly
    plt.gca().xaxis.set_major_locator(mdates.MonthLocator())
    plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
    plt.gcf().autofmt_xdate()

    plt.show()
```

```python
# Remove currency symbol and convert 'Order Amount' to numeric
df_clv['Order Amount'] = df_clv['Order Amount'].str.replace('MYR ', '').astype(float)
```

```python
# Calculate total revenue, number of orders, and unique customers
total_revenue = df_clv['Order Amount'].sum()
total_orders = df_clv['Order Amount'].count()
unique_customers = df_clv['Buyer Username'].nunique()
```

```python
# Calculate Average Purchase Value (APV)
apv = total_revenue / total_orders

# Calculate Average Purchase Frequency Rate (APFR)
apfr = total_orders / unique_customers

# Calculate Customer Value (CV)
cv = apv * apfr

# Estimate Average Customer Lifespan (ACL)
# Assuming ACL in months as the difference between first and last order date for each customer
df_clv['Month'] = df_clv['Created Time'].dt.to_period('M')
customer_lifespan = df_clv.groupby('Buyer Username')['Month'].nunique().mean()

# Calculate Customer Lifetime Value (CLV)
clv = cv * customer_lifespan
```

```python
# Print results
print(f"Total Revenue: MYR {total_revenue:.2f}")
print(f"Total Orders: {total_orders}")
print(f"Unique Customers: {unique_customers}")
print(f"Average Purchase Value (APV): MYR {apv:.2f}")
print(f"Average Purchase Frequency Rate (APFR): {apfr:.2f} orders per customer")
print(f"Customer Value (CV): MYR {cv:.2f}")
print(f"Average Customer Lifespan (ACL): {customer_lifespan:.2f} months")
print(f"Customer Lifetime Value (CLV): MYR {clv:.2f}")
```
